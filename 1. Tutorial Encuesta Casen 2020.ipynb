```python
casen_biobio = casen_hogar[casen_hogar["region"] == "Región del Biobío"]
casen_biobio.head()
```

```python
# tot_per
# Total de personas en el hogar
```

```python
fig, ax = plt.subplots(figsize=(20, 5))

sns.barplot(
    x="comuna",
    y="tot_per",
    data=casen_biobio
)
```

```python
comunas_segun_tot_per = casen_biobio.groupby("comuna").agg({"tot_per": "mean"}).sort_values("tot_per")
comunas_segun_tot_per.head()
```

```python
comunas_palette = casen_biobio.groupby(["comuna", "provincia"]).agg({"tot_per": "mean"}).sort_values("tot_per")
comunas_palette = comunas_palette.reset_index()
comunas_palette.head()
```

```python
casen_biobio.provincia.unique()
```

```python
sns.set(font_scale=2.25)

fig, ax = plt.subplots(figsize=(20, 5))

palette = {
    "Concepción": "blue",
    "Arauco": "orange",
    "Biobío": "brown"
}

sns.barplot(
    x="comuna",
    y="tot_per",
    data=casen_biobio.dropna(subset=["comuna", "tot_per"]),
    palette=comunas_palette["provincia"].replace(palette),
    order=comunas_segun_tot_per.index
)

ax.tick_params(axis="x", rotation=90)

ax.set_xlabel("Comuna")
ax.set_ylabel("Personas por Hogar")

sns.despine()
```

```python
comunas_segun_ytotcorh = casen_biobio.groupby("comuna").agg({"ytotcorh": "mean"}).sort_values("ytotcorh")
comunas_segun_ytotcorh.head()

comunas_palette = casen_biobio.groupby(["comuna", "provincia"]).agg({"ytotcorh": "mean"}).sort_values("ytotcorh")
comunas_palette = comunas_palette.reset_index()
comunas_palette.head()
```

```python
sns.set(font_scale=2.25)

fig, ax = plt.subplots(figsize=(20, 5))

palette = {
    "Concepción": "blue",
    "Arauco": "orange",
    "Biobío": "brown"
}

sns.barplot(
    x="comuna",
    y="ytotcorh",
    data=casen_biobio.dropna(subset=["comuna", "ytotcorh"]),
    palette=comunas_palette["provincia"].replace(palette),
    order=comunas_segun_ytotcorh.index
)

ax.tick_params(axis="x", rotation=90)

ax.set_xlabel("Comuna")
ax.set_ylabel("Ingreso por Hogar")

sns.despine()
```

```python
sns.set(font_scale=2.25)

fig, ax = plt.subplots(figsize=(20, 5))

comunas_segun_ytotcorh = casen_biobio.groupby("comuna").agg({"ytotcorh": "median"}).sort_values("ytotcorh")
comunas_segun_ytotcorh.head()

comunas_palette = casen_biobio.groupby(["comuna", "provincia"]).agg({"ytotcorh": "median"}).sort_values("ytotcorh")
comunas_palette = comunas_palette.reset_index()
comunas_palette.head()

palette = {
    "Concepción": "blue",
    "Arauco": "orange",
    "Biobío": "brown"
}

sns.barplot(
    x="comuna",
    y="ytotcorh",
    data=casen_biobio.dropna(subset=["comuna", "ytotcorh"]),
    palette=comunas_palette["provincia"].replace(palette),
    order=comunas_segun_ytotcorh.index,
    estimator=np.median
)

ax.tick_params(axis="x", rotation=90)

ax.set_xlabel("Comuna")
ax.set_ylabel("Ingresos Hogar (Mediana)")

sns.despine()
```

```python
# y26d_hog. Últimos 12 meses,
# ¿alguien recibió Ingreso Familiar de
# Emergencia? 
```

```python
data = casen_concepcion.groupby(["comuna", "y26d_hog"]).agg({"folio": "count"})
data["rate"] = data["folio"].groupby(level=[0], group_keys=False).apply(lambda x: x/x.sum())

data = data.reset_index()
data = data[data["y26d_hog"] != "No sabe"]
data["y26d_hog"] = data["y26d_hog"].cat.remove_unused_categories()

fig, ax = plt.subplots(figsize=(20, 5))

sns.barplot(
    x="comuna",
    y="rate",
    hue="y26d_hog",
    data=data,
)

ax.tick_params(axis="x", rotation=90)
ax.set_ylabel("Nro. Hogares")
ax.set_xlabel("Comuna")
ax.legend(title="")
```

```python
# data = data[data["y26d_hog"] != "No sabe"]
data2 = casen_biobio[casen_biobio["y26d_hog"] != "No sabe"].groupby(["provincia", "comuna", "y26d_hog"]).agg({"folio": "count"})
data2["rate"] = data2["folio"].groupby(level=[0, 1], group_keys=False).apply(lambda x: x/x.sum())

data2 = data2.reset_index()
data2["y26d_hog"] = data2["y26d_hog"].cat.remove_unused_categories()
data2 = data2[data2["y26d_hog"] == "Sí"].dropna()

data2 = data2.sort_values("rate")

palette = {
    "Concepción": "blue",
    "Arauco": "orange",
    "Biobío": "brown"
}
fig, ax = plt.subplots(figsize=(20, 5))

sns.barplot(
    x="comuna",
    y="rate",
    palette=data2["provincia"].replace(palette),
    data=data2,
)

ax.tick_params(axis="x", rotation=90)
ax.set_yticklabels([f"{int(x * 100)}%" for x in ax.get_yticks()])
ax.set_ylabel("% Hogares")
ax.set_xlabel("Comuna")
ax.legend(title="")
```
